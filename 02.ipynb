# Data exploration

To better understand and interpret the results of various metrices applied on our network, we decided to see the distribution of various attribute's values and what are the average values. 

Let's start by importing the csv data of the two weeks of April 07 and 08.

#### Distribution of `CATEGORY` of videos.

```python
#distribution_Frequency(category_values,'category of video')

def plot_categorical_frequency(data, label):
    # Extract the categorical variable from the DataFrame
    categorical_variable = data

    # Get unique categories and their counts
    unique_categories, counts = categorical_variable.value_counts().index, categorical_variable.value_counts().values

    # Set the figure size
    plt.figure(figsize=(8, 6))  # Adjust the width and height as desired

    # Create bar chart
    plt.bar(unique_categories, counts)

    # Add labels and title
    plt.xlabel(label)
    plt.ylabel('Frequency')
    plt.title('Frequency of Categorical Variable')

    # Rotate x-axis labels if needed
    plt.xticks(rotation=75)  # Adjust the rotation angle as desired

    # Display the plot
    plt.show()

plot_categorical_frequency(category_values,'category of video')
```

#### Distribution of `RATING` of videos.

```python
distribution_Frequency(rate_values,'rate of video')
```

#### Distribution of `RATING count` of videos.

```python
#distribution_Frequency(ratings_values,'rate count of video')

for i in range(len(rate_values)):
    if pd.isna(rate_values[i]) or rate_values[i] is None:
        rate_values[i] = 0
        
for i in range(len(ratings_values)):
    if pd.isna(ratings_values[i]) or ratings_values[i] is None:
        ratings_values[i] = 0

fig, axs = plt.subplots(2, 1, figsize=(20, 10), sharex=True, sharey=True,
                        tight_layout=True)

# We can increase the number of bins on each axis
#axs[0].hist2d(rate_values, ratings_values, bins=40)

# As well as define normalization of the colors
axs[0].hist2d(rate_values, ratings_values, bins=40, norm=colors.LogNorm())

# We can also define custom numbers of bins for each axis
axs[1].hist2d(rate_values, ratings_values, bins=(80, 20), norm=colors.LogNorm())

plt.show()
```
